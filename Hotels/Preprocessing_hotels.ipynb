```python
df=df[df[' countyName']=='United States']
df = df.reset_index(drop=True)
df.head()
```

```python
df2 = df.copy()
split_data = df[' cityName'].str.split(',', expand=True)
df['city'] = split_data[0]
df['state'] = split_data[1]
```

```python
df = df.drop(columns=[' cityName'])
df.head()
```

```python
states_to_drop = ["Hawaii", "Hawaii County", "Alaska"]

df['state'] = df['state'].str.strip().str.lower()

states_to_drop = [state.lower() for state in states_to_drop]

df = df[~df['state'].isin(states_to_drop)]

df.reset_index(drop=True, inplace=True)
```

```python
df.columns = df.columns.str.strip()

df=df.drop(columns=['countyCode','cityCode','HotelCode','Attractions','Description','FaxNumber','PinCode'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Split latitude and longitude from 'Map'
df[['latitude', 'longitude']] = df['Map'].str.split('|', expand=True).astype(float)

# Preprocess - scale the location data
scaler = StandardScaler()
df[['latitude', 'longitude']] = scaler.fit_transform(df[['latitude', 'longitude']])
```

```python
df = df.drop(columns=['Map'])
```

```python
rating_map = {'OneStar': 1, 'TwoStar': 2, 'ThreeStar': 3, 'FourStar': 4, 'FiveStar': 5}
df['HotelRatingNumeric'] = df['HotelRating'].map(rating_map)

df.drop('HotelRating', axis=1, inplace=True)
```

```python
df.rename(columns={'HotelRatingNumeric': 'HotelRating'}, inplace=True)
```

```python
# Remove rows with NaN in 'Address' column, modifying the original dataframe
df.dropna(subset=['longitude', 'latitude', 'Address', 'HotelFacilities'], inplace=True)
```

```python
print(df.isna().sum())
```

```python
import re

# Step 1: Remove specific non-digit characters (including parentheses)
df['PhoneNumber'] = df['PhoneNumber'].fillna('').str.replace(r'[+\-\s\(\)]', '', regex=True)

# Step 2: Clean individual phone numbers
def clean_phone_number(phone):
    if len(phone) == 11 and phone[0] == '1':
        # Remove the leading 1 if the phone number is exactly 11 digits long
        phone = phone[1:]
    # If the phone number is less than 10 digits, return an empty string
    return phone if len(phone) == 10 else ''

# Step 3: Process multiple phone numbers split by '|'
def process_phone_numbers(phone_numbers):
    # Split the phone numbers by '|' if there are multiple, else treat as a single number
    phone_list = phone_numbers.split('|')
    # Clean each phone number and keep only valid ones
    cleaned_numbers = [clean_phone_number(re.sub(r'\D', '', num)) for num in phone_list]
    # Filter out empty strings (invalid phone numbers) and join valid numbers with a comma
    return ', '.join(filter(None, cleaned_numbers))

# Step 4: Apply the function to the PhoneNumber column
df['PhoneNumber'] = df['PhoneNumber'].apply(process_phone_numbers)
```

```python
base_url = "https://www.booking.com/hotel/us/"

# Replace NaN values in 'HotelWebsiteUrl' with a formatted URL that uses the hotel name
df['HotelWebsiteUrl'] = df['HotelWebsiteUrl'].fillna(
    df['HotelName'].str.lower()  # Convert the hotel name to lowercase
    .str.replace(' ', '-', regex=True)  # Replace spaces with hyphens
    .str.replace(r'[^a-zA-Z0-9\-]', '', regex=True)  # Remove non-alphanumeric characters
    .apply(lambda name: f"{base_url}{name}.htm")  # Add to base URL format
)

# Check the result
df.head()
```
